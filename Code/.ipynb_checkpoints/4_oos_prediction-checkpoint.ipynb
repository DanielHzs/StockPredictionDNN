```python
import os
import tensorflow as tf
import pandas as pd
import numpy as np
from pathlib import Path
import train_ensemble

CONFIG = train_ensemble.CONFIG
make_dataset = train_ensemble.make_dataset

CONFIG['label'] = 'ret_exc_lead1m'
CONFIG['weight'] = 'w_ew'
CONFIG['model_dir'] = '/gpfs/home/zh283/StockPredictionDNN/Code/chkpts_script'
# ─── 1. Locate & load your ensemble ─────────────────────────────────────────────
MODEL_DIR = Path(CONFIG.get("model_dir", "/gpfs/home/zh283/StockPredictionDNN/chkpts_script"))
model_paths = sorted(MODEL_DIR.glob("ensemble_*.keras"))
print(f"Found {len(model_paths)} models:", model_paths)

ensemble = [tf.keras.models.load_model(str(p)) for p in model_paths]

# ─── 1. Load prediction data as before ──────────────────────────────────────────
CONFIG['tfrecord_dir'] = os.path.join(CONFIG['data_dir'], 'tfrecords', 'pct')
pred_ds = make_dataset(CONFIG["predict_years"], mode="predict")

feat_list, label_list, weight_list, meta_buf = [], [], [], {}
for feats, labels, weights, meta in pred_ds:
    feat_list.append(feats.numpy())
    label_list.append(labels.numpy().ravel())
    weight_list.append(weights.numpy().ravel())
    for k, v in meta.items():
        meta_buf.setdefault(k, []).append(v.numpy())

X = np.concatenate(feat_list, axis=0)
y_true = np.concatenate(label_list, axis=0)
wts    = np.concatenate(weight_list, axis=0)
meta_np = {k: np.concatenate(v, axis=0) for k, v in meta_buf.items()}

print(f"Data loaded: {X.shape[0]} examples.")

# ─── 2. Get all model predictions in one go ────────────────────────────────────
print("Running ensemble predictions in‐memory…")
all_preds = np.stack([
    model.predict(X, batch_size=CONFIG['eval_batch_size']).ravel()
    for model in ensemble
], axis=0)  # shape (n_models, N)

# ─── 3. Build one DataFrame per model + one for the mean ──────────────────────
print("Assembling per‑model DataFrames…")
dfs = []

# individual models
for idx in range(all_preds.shape[0]):
    df_i = pd.DataFrame({
        "label":      y_true,
        "prediction": all_preds[idx],  # model idx’s preds
        "weight":     wts,
        **meta_np
    })
    dfs.append(df_i)

# ensemble mean
print("Assembling ensemble‑mean DataFrame…")
y_pred_mean = all_preds.mean(axis=0)
df_ensemble = pd.DataFrame({
    "label":      y_true,
    "prediction": y_pred_mean,
    "weight":     wts,
    **meta_np
})
dfs.append(df_ensemble)

for df in dfs:
    df['eom'] = pd.to_datetime(df['eom'].astype(str), format='%Y%m%d')
```

```python
def construct_quantile_portfolios(df,
                                  signal_col='prediction',
                                  date_col='eom',
                                  weight_col='weight',
                                  return_col='label',
                                  n=10):
    """
    Build n quantile portfolios (value‐weighted) plus a long‐short (top minus bottom).
    
    Parameters
    ----------
    df : pd.DataFrame
      Must contain at least [date_col, weight_col, return_col, signal_col].
    date_col : str
      Column name to group on (e.g. 'eom').
    weight_col : str
      Column with stock weights (used for value‐weighting).
    return_col : str
      Column with the per‐stock returns.
    signal_col : str
      Column with the signal used to sort into quantiles.
    n : int
      Number of quantiles (e.g. 10 for deciles).
    
    Returns
    -------
    port_rets : pd.DataFrame
      Indexed by unique values of date_col (sorted). Columns are:
        'q1', 'q2', …, f'q{n}', 'long_short'
      where q1 is the lowest‐signal bucket and qn the highest, and
      long_short = qn − q1.
    """
    # 1) Drop rows with missing key fields
    df0 = df.dropna(
        subset=[date_col, weight_col, return_col, signal_col]).copy()

    # 2) Compute in‐group percentile rank of signal
    #    rank_pct ∈ (0,1], then quantile ∈ {1,…,n}
    df0['rank_pct'] = df0.groupby(date_col)[signal_col]\
                         .rank(method='first', pct=True)
    df0['quantile'] = np.ceil(df0['rank_pct'] * n).astype(int).clip(1, n)

    # 3) Compute weighted returns per (date,quantile) in one go
    #    create a helper column wR = weight * return
    df0['wR'] = df0[weight_col] * df0[return_col]

    grp = df0.groupby([date_col, 'quantile'])
    sum_wr = grp['wR'].sum()  # weighted return sum
    sum_w = grp[weight_col].sum()  # weight sum

    # 4) Pivot to wide: rows = dates, cols = quantiles
    port = (sum_wr / sum_w).unstack(fill_value=np.nan)
    # name columns q1…qn
    port.columns = [f"q{q}" for q in port.columns]

    # 5) Long‐short spread
    port['long_short'] = port[f"q{n}"] - port["q1"]

    # 6) ensure sorted by date
    port = port.sort_index()
    return port


def evaluate_portfolios(returns, freq=12):
    """
    Evaluate performance of one or more portfolio return series.
    
    Parameters
    ----------
    returns : pd.DataFrame
        Periodic returns, indexed by date, columns = portfolio names.
    freq : int, optional
        Number of periods per year (default 252 for daily returns).
    rf : float, optional
        Annualized risk‐free rate (default 0.0). Must be in same units as freq,
        so per‐period rf_per = (1+rf)**(1/freq) - 1.
    
    Returns
    -------
    perf : pd.DataFrame
        Indexed by metric name, columns = portfolio names, containing:
          - mean_return : annualized average return
          - vol         : annualized volatility
          - sharpe      : annualized Sharpe ratio
          - max_drawdown: maximum drawdown (as positive number, e.g. 0.25 = 25%)
    """
    # Drop any missing data
    rets = returns.dropna(how='all')

    # 1) annualized mean return
    mean_ret = rets.mean() * freq

    # 2) annualized volatility
    vol = rets.std(ddof=1) * np.sqrt(freq)

    # 3) Sharpe ratio
    # subtract per‐period rf, scale by sqrt(freq)
    sharpe = mean_ret / vol

    # 4) Maximum drawdown
    # compute cumulative wealth index
    wealth_index = (1 + rets).cumprod()
    running_max = wealth_index.cummax()
    drawdown = 1 - wealth_index / running_max
    max_dd = drawdown.max()

    perf = pd.DataFrame({
        "mean_return": mean_ret,
        "vol": vol,
        "sharpe": sharpe,
        "max_drawdown": max_dd
    })

    return perf  # metrics as rows, portfolios as columns

import matplotlib.pyplot as plt
import numpy as np

def plot_portfolio_cumulative_returns(returns_df):
    """
    Plots cumulative and log cumulative returns for portfolios.
    
    Parameters:
    -----------
    returns_df : pd.DataFrame
        Periodic returns indexed by date, columns are portfolio names.
    """
    # Compute cumulative returns and log cumulative returns
    cum_returns = (1 + returns_df).cumprod()
    log_cum_returns = np.log(cum_returns)

    # Plot cumulative returns
    plt.figure()
    for col in cum_returns.columns:
        plt.plot(cum_returns.index, cum_returns[col], label=col)
    plt.xlabel('Date')
    plt.ylabel('Cumulative Return')
    plt.title('Cumulative Returns of Portfolios')
    plt.legend()
    plt.show()

    # Plot log cumulative returns
    plt.figure()
    for col in log_cum_returns.columns:
        plt.plot(log_cum_returns.index, log_cum_returns[col], label=col)
    plt.xlabel('Date')
    plt.ylabel('Log Cumulative Return')
    plt.title('Log Cumulative Returns of Portfolios')
    plt.legend()
    plt.show()
```

```python
qret_list = []
perf_list = []
for df in dfs:
    qret_list.append(construct_quantile_portfolios(df))
    perf_list.append(evaluate_portfolios(qret_list[-1]))
```

```python
for perf in perf_list:
    print(perf)
```
